```python
#Making a pandas DataFrame
Df = pd.read_csv(train_path,names=Column_Name,header=None)
Df.head()
```

Data-Preprocessing

```python
#Removing the Feature ['difficulty_level]
Df = Df.drop(['difficulty_level'],axis=1)
```

```python
#Categoriziing these attacks into 4 main types of Network Attacks
def change_label(df):
  df.label.replace(['apache2','back','land','neptune','mailbomb','pod','processtable','smurf','teardrop','udpstorm','worm'],'Dos',inplace=True)
  df.label.replace(['ftp_write','guess_passwd','httptunnel','imap','multihop','named','phf','sendmail',
       'snmpgetattack','snmpguess','spy','warezclient','warezmaster','xlock','xsnoop'],'R2L',inplace=True)
  df.label.replace(['ipsweep','mscan','nmap','portsweep','saint','satan'],'Probe',inplace=True)
  df.label.replace(['buffer_overflow','loadmodule','perl','ps','rootkit','sqlattack','xterm'],'U2R',inplace=True)
  return df
```

```python
Df = change_label(Df)
Df['label'].value_counts()
```

```python
Df['service'].unique()
```

Normalizing the Data

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
```

```python
#A Function to normalize our Dataset
def normalization(df,col):
    for i in col:
        arr = df[i]
        arr = np.array(arr)
        df[i] = Scaler.fit_transform(arr.reshape(len(arr),1))
    return df
```

```python
numeric_col = Df.select_dtypes(include='number').columns
Df = normalization(Df.copy(),numeric_col)
```

```python
Df.head()
```

```python
#using One-Hot encoder for these Features
cat_col = ['protocol_type','service','flag']
#Creating a Separate DataFrame for these Features
categorical = Df[cat_col]
```

```python
#One Hot-Encoded Data
categorical = pd.get_dummies(categorical,columns=cat_col)
categorical.head()
```

```python
Copy = Df
```

```python
Copy = pd.concat([Copy,categorical],axis=1)
Copy.head()
```

```python
Copy = Copy.drop(['protocol_type','service','flag'],axis=1)
```

```python
Copy.head()
```

```python
Copy['label'].unique()
```

```python
#Converting the whole Dataset into Binary Classification Dataset
Copy['label'] = pd.DataFrame(Copy.label.map(lambda x:'normal' if x=='normal' else 'abnormal'))
Copy['label'].unique()
```

```python
Copy.head()
```

```python
Copy['label'] = pd.DataFrame(Copy.label.map(lambda x:0 if x=='normal' else 1))
Copy['label'].unique()
```

```python
Df_x = Copy.drop(['label'],axis=1)
Df_y = Copy['label']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(Df_x,Df_y,test_size=0.25,random_state=42 )
```

```python
from sklearn.linear_model import LogisticRegression
Model = LogisticRegression()
```

```python
Model.fit(train_x, train_y)
```

```python
Predict = Model.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy :" ,accuracy_score(test_y,Predict)*100)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras import layers
```

```python
Deep_Model = Sequential()
```

```python
#First Layer of Neuron
Deep_Model.add(layers.Dense(122,input_dim=122))
```

```python
#Second Layer
Deep_Model.add(layers.Dense(500,activation='relu'))
```

```python
#Third Layer
Deep_Model.add(layers.Dense(500,activation='relu'))
```

```python
#3rd Neural Layer
Deep_Model.add(layers.Dense(100,activation='relu'))
```

```python
#Final layer
Deep_Model.add(layers.Dense(1,activation='sigmoid'))
```

```python
Deep_Model.summary()
```

```python
_, val_x, _, val_y = train_test_split(train_x,train_y,test_size=0.25,random_state=42)
```

```python
Deep_Model.compile(optimizer = tf.keras.optimizers.RMSprop(learning_rate= 2e-5) , loss='binary_crossentropy',metrics='accuracy')
```

```python
Deep_Model.fit(train_x,train_y,epochs=10,validation_data=(val_x,val_y),validation_steps=10)
```

```python
Predict = Deep_Model.predict(test_x)
```

```python
Predict = np.round(Predict)
```

```python
print("Accuracy of Deep Model: ",accuracy_score(test_y,Predict)*100)
```
